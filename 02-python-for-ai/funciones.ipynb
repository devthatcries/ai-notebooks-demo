1. Cree un programa que calcule la desviación estandar de una muestra de datos

```python
def calcular_desviacion_estandar(valores):
    n = len(valores)
    if n == 0 or n == 1:
        return -1

    # Calcular la media
    suma = 0
    for valor in valores:
        suma += valor
    media = suma / n

    # Calcular la suma de las diferencias al cuadrado
    suma_diferencias_cuadrado = 0
    for valor in valores:
        diferencia = valor - media
        suma_diferencias_cuadrado += diferencia ** 2

    # Calcular la varianza
    varianza = suma_diferencias_cuadrado / (n-1)

    # Calcular la desviación estándar
    desviacion_estandar = varianza ** 0.5

    return desviacion_estandar
```

```python
# Ejemplo de uso
valores = [10, 12, 23, 23, 16, 23, 21, 16]
desviacion = calcular_desviacion_estandar(valores)
```

```python
if desviacion == -1:
  print("Has enviado una lista vacía o con un solo elemento")
else:
  print(f"La desviación estándar es: {round(desviacion,2)}")
```

2. Cree un programa que convierta una determinada cantidad de dinero de euros a dólares estadounidenses (USD) y viceversa

```python
DOLAR_A_EURO = 0.91
EURO_A_DOLAR = 1 / DOLAR_A_EURO
```

```python
def convertir_moneda(cantidad, moneda_origen, moneda_destino):
    if moneda_origen == 'dolar' and moneda_destino == 'euro':
        return cantidad * DOLAR_A_EURO
    elif moneda_origen == 'euro' and moneda_destino == 'dolar':
        return cantidad * EURO_A_DOLAR
    else:
        return None  # Conversión no soportada
```

```python
cantidad = float(input("Ingrese la cantidad a convertir: "))
moneda_origen = input("Ingrese la moneda de origen (dólar/euro): ").lower()
moneda_destino = input("Ingrese la moneda de destino (dólar/euro): ").lower()
```

```python
resultado = convertir_moneda(cantidad, moneda_origen, moneda_destino)
if resultado is not None:
    print(f"{round(cantidad,2)} {moneda_origen} son {round(resultado,2)} {moneda_destino}.")
else:
    print("Conversión no soportada.")
```

3. Cree un programa que calcule el producto punto entre dos vectores

```python
def producto_punto(vector1, vector2):
    if len(vector1) != len(vector2):
        return -1

    producto = 0
    for i in range(len(vector1)):
        producto += vector1[i] * vector2[i]

    return producto
```

```python
# Ejemplo de uso
vector1 = [1, 2, 3]
vector2 = [4, 5, 6]
resultado = producto_punto(vector1, vector2)
```

```python
if resultado == -1:
  print("Vectores tienen distinta dimension")
else:
  print(f"El producto punto entre los vectores es: {resultado}")
```
