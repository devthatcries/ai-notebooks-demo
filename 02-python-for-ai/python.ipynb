<p>Calcular el volumen de un cono</p>

```python
PI = 3.1415
r = 7
h = 10
volumen = (PI*(r**2)*h)/3
volumen
```

<p>Indique si un numero es primo o no</p>

```python
NUMERO = 10
i = 2
es_primo = True
while i<NUMERO:
  if NUMERO%i==0:
    es_primo = False
    break
  i+=1

if es_primo == True:
  print(f'El numero {NUMERO} es primo')
else:
  print(f'El numero {NUMERO} no es primo')
```
