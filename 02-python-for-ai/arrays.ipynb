```python
import numpy as np
```

```python
# Sumar elementos de dos arrays de forma independiente
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
sum_result = np.add(a, b)
sum_result
```

```python
# Multiplicar elementos de dos arrays de forma independiente
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
prod_result = np.multiply(a, b)
prod_result
```

```python
# Sumar todos los elementos de un array
a = np.array([1, 2, 3, 4, 5])
total_sum = np.sum(a)
total_sum
```

```python
# Sumar todos los elementos de un array bidimensional
b = np.array([[1, 2, 3], [4, 5, 6]])
total_sum = np.sum(b)
total_sum
```
