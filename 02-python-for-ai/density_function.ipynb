```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función a integrar
def func(x):
    return np.sin(x)  # Ejemplo de función: seno
```

```python
# Definir los límites de integración
a = 0
b = np.pi
n = 1000  # Número de puntos
```

```python
# Generar los puntos de evaluación
x = np.linspace(a, b, n)
y = func(x)
```

```python

# Calcular el área bajo la curva usando la regla del trapecio
dx = (b - a) / (n - 1)
area = np.sum((y[:-1] + y[1:]) * dx / 2)

print(f'Área bajo la curva: {round(area,2)}')
```

```python
# Graficar la función y el área bajo la curva
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = sin(x)')
plt.fill_between(x, y, alpha=0.3, label='Área bajo la curva')
plt.title('Área bajo la curva de f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
