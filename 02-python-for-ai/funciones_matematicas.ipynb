```python
import numpy as np
```

```python
# Producto punto
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_result = np.dot(a, b)
dot_result
```

```python
# Raiz cuadrada
a = np.array([1, 4, 9, 16])
sqrt_result = np.sqrt(a)
sqrt_result
```

```python
# Seno y coseno
angles = np.array([0, np.pi/2, np.pi])
sin_result = np.sin(angles)
cos_result = np.cos(angles)
print(sin_result)
print(cos_result)
```

```python
# Media
data = np.array([1, 2, 3, 4, 5])
mean_result = np.mean(data)
mean_result
```

```python
# Mediana
data = np.array([1, 2, 3, 4, 5])
median_result = np.median(data)
median_result
```

```python
# Varianza
data = np.array([1, 2, 3, 4, 5])
var_result = np.var(data)
var_result
```

```python
# Desviacion estandar
data = np.array([1, 2, 3, 4, 5])
std_result = np.std(data)
std_result
```

```python
# Correlacion
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
corr_result = np.corrcoef(x, y)
corr_result
```
