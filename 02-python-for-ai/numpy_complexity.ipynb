```python
import time
import random
```

```python
vector1 = [random.randint(0, 100) for _ in range(10000000)]
vector2 = [random.randint(0, 100) for _ in range(10000000)]
```

<p>Python</p>

```python
def producto_punto(vector1, vector2):
    if len(vector1) != len(vector2):
        return -1

    inicio = time.time()

    producto = 0
    for i in range(len(vector1)):
        producto += vector1[i] * vector2[i]

    fin = time.time()

    tiempo = fin - inicio

    return producto, tiempo
```

```python
# Ejemplo de uso
resultado, tiempo = producto_punto(vector1, vector2)
```

```python
if resultado == -1:
  print("Vectores tienen distinta dimension")
else:
  print(f"El producto punto entre los vectores es: {resultado}")
  print(f"El tiempo de procesamiento fue de: {round(tiempo,6)} segundos")
```

<p>NumPy</p>

```python
import numpy as np
```

```python
vector1 = np.array(vector1)
vector2 = np.array(vector2)
```

```python
print(vector1.shape)
print(vector2.shape)
```

```python
def producto_punto_np(vector1, vector2):
    if vector1.shape[0] != vector2.shape[0]:
       return -1

    inicio = time.time()

    producto = np.dot(vector1, vector2)

    fin = time.time()

    tiempo = fin - inicio

    return producto, tiempo
```

```python
# Ejemplo de uso
resultado, tiempo = producto_punto_np(vector1, vector2)
```

```python
if resultado == -1:
  print("Vectores tienen distinta dimension")
else:
  print(f"El producto punto entre los vectores es: {resultado}")
  print(f"El tiempo de procesamiento fue de: {round(tiempo,6)} segundos")
```
