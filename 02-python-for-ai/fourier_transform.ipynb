```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Parámetros del pulso rectangular
T = 2.0      # Duración del pulso
fs = 1000    # Frecuencia de muestreo
t = np.arange(-2.0, 2.0, 1/fs) # Intervalo de tiempo
```

```python
# Generar el pulso rectangular
pulse = np.where(np.abs(t) <= T/2, 1.0, 0.0)
```

```python
# Calcular la transformada de Fourier
freqs = np.fft.fftfreq(len(t), 1/fs)
fft_pulse = np.fft.fft(pulse)
```

```python
# Magnitud de la transformada de Fourier
magnitude = np.abs(fft_pulse)
```

```python
# Graficar el pulso rectangular
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, pulse, label='Pulso Rectangular')
plt.title('Pulso Rectangular')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

# Graficar la transformada de Fourier
plt.subplot(2, 1, 2)
plt.plot(freqs, magnitude, label='Transformada de Fourier')
plt.title('Transformada de Fourier del Pulso Rectangular')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
